```python
class University:
    def __init__(self,name,location):
        self.name=name
        self.location=location
    def display(self):
        print("Name is "+ self.name + " location is "+ self.location)
        print(f"Name is {self.name} and location is {self.location}")
        
class Department(University):
    def __init__(self,name,program,uname,ulocation):
        super().__init__(uname,ulocation)
        self.name=name
        self.program=program
    def display(self):
        super().display()
        print(f"The name is {self.name} and program are {self.program}")
        
```

```python
def getCount(str_list):
    count=0
    
    for i,value in enumerate(str_list):
        if len(value)>=2 and value[0].lower()==value[len(value)-1].lower():
            count=count+1
    return count
```

```python
str_list=["Nust is in Islamabad", "My aim","That boy"]
count=(getCount(str_list))
count
```

### Task 2

```python
def sort_list(us_list):
    x_list=[]
    ot_list=[]
    ot_ind_list=[]
    final_list=[]
    unique_list=[]
    count=0
    for i,value in enumerate(us_list):
        if value[0]=="x":
            x_list.append(i)
        else:
            ot_list.append(value)
            ot_ind_list.append(i)
            
            
    ot_list.sort()
    for x in ot_list:
        if x not in unique_list:
            unique_list.append(x)
            
            
    for i in range(len(us_list)):
        if i in x_list:
            final_list.append(us_list[i])
        elif i in ot_ind_list:
            final_list.append(ot_list[count])
            count=count+1
    
    return final_list
```

```python
us_list=["zahid","xli","xavier","ali","abdullah","muhammad"]
sort_list(us_list)
```

### Task 3

```python
def sort_tup(li_tup):
    end=[]
    end_index=[]

    res=[]
    for i,value in enumerate(li_tup):
        end.append([value[len(value)-1],i]) 
    end.sort()
    for i in end:
        end_index.append(i[1])

    for i in range(len(li_tup)):
        res.append(li_tup[end_index[i]])
    return res
```

```python
li_tup=[(1,2,3,4),(3,6,8),(2,3)]

sort_tup(li_tup)
```

```python
def sort_merge(sort_li_1,sort_li_2):
    merge=sort_li_1+sort_li_2
    merge.sort()
    return merge
```

```python
sort_li_1=[2,3,4,5,6]
sort_li_2=[10,2,1,17]

sort_merge(sort_li_1,sort_li_2)
```
